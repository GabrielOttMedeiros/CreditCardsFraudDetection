```python
import pandas as pd

lr = pd.read_csv('LR_results.csv')
ada_dtc = pd.read_csv('ADA_DTC.csv')
dtc = pd.read_csv('DTC_results.csv')
nn = pd.read_csv('NN_results.csv')
rf = pd.read_csv('RF_results.csv')
svc = pd.read_csv('SVC_results.csv')
gbc = pd.read_csv('GBC_results.csv')


lr['Model'] = 'LogisticRegression'
ada_dtc['Model'] = 'AdaBoostDecisionTrees'+ada_dtc['max_depth'].astype(str)+'_'+ada_dtc['estimators'].astype(str)+'_'+ada_dtc['learning_rate']
nn['Model'] = 'NeuralNetworks'+nn['number_of_neurons'].astype(str)+'_'+nn['activation'].astype(str)+'_'+nn['activation2']
svc['Model'] = 'SupportVectorMachine'+svc['max_depth'].astype(str)+'_'+svc['estimators'].astype(str)+'_'+svc['learning_rate']
dtc['Model'] = 'DecisionTrees'+dtc['max_depth'].astype(str)+'_'+dtc['estimators'].astype(str)+'_'+dtc['learning_rate']
rf['Model'] = 'RandomForest'+rf['max_depth'].astype(str)+'_'+rf['estimators'].astype(str)+'_'+rf['learning_rate']
gbc['Model'] = 'GradientBoosting'+gbc['max_depth'].astype(str)+'_'+gbc['estimators'].astype(str)+'_'+gbc['learning_rate']



lr = lr[['accuracy','recall','Performance','cut_off','Model']]
ada_dtc = ada_dtc[['accuracy','recall','Performance','cut_off','Model']]
nn = nn[['accuracy','recall','Performance','Model']]
svc = svc[['accuracy','recall','Performance','cut_off','Model']]
dtc = dtc[['accuracy','recall','Performance','cut_off','Model']]
rf = rf[['accuracy','recall','Performance','cut_off','Model']]
gbc = gbc[['accuracy','recall','Performance','cut_off','Model']]



results = pd.concat([lr,ada_dtc,nn, svc, dtc, rf, gbc])
```
